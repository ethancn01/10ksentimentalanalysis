```python
import pandas as pd
import numpy as np
```

```python
financial_data=pd.read_csv('financial_data.csv')
financial_data=financial_data.iloc[:,1:]
```

```python
nan_values = financial_data.isna().sum()

# Display the count of NaN values for each column
print(nan_values)
```

```python
financial_data
```

```python
financial_data['CIK'] = financial_data['CIK'].astype(str).str.zfill(10)
financial_data = financial_data.astype(str)
```

```python
roe=pd.read_excel('ROE.xlsx')
asset=pd.read_excel('assets.xlsx')
```

```python
roe
```

```python
asset
```

```python
year_column_roe = roe.columns[1:]
year_column_asset=asset.columns[1:]
```

```python
roe_melted = pd.melt(roe, id_vars=['Ticker','CIK'], value_vars = year_column_roe, var_name = 'Year')
asset_melted = pd.melt(asset, id_vars=['Ticker','CIK'], value_vars = year_column_asset,  var_name = 'Year')
```

```python
roe_melted.columns
```

```python
roe_melted = roe_melted.astype(str)
roe_melted['CIK'] = roe_melted['CIK'].astype(str).str.zfill(10)
#roe_melted.columns = roe_melted.columns.str.upper()
roe_melted.rename(columns={'value':'ROE'},inplace=True)

asset_melted = asset_melted.astype(str)
asset_melted['CIK'] = asset_melted['CIK'].astype(str).str.zfill(10)
#asset_melted.columns = asset_melted.columns.str.upper()
asset_melted.rename(columns={'value':'Asset'},inplace=True)
```

```python
roe_melted = roe_melted.sort_values(by=['Ticker','Year'])
asset_melted = asset_melted.sort_values(by=['Ticker','Year'])
```

```python
asset_melted.columns
```

```python
asset_melted['Y-1 Asset'] = asset_melted.groupby('CIK')['Asset'].shift(1)
asset_melted['Y+1 Asset'] = asset_melted.groupby('CIK')['Asset'].shift(-1)
```

```python
roe_melted.columns
```

```python
# Group by CIK and then use the shift function to get the lagged and lead values
roe_melted['Y-1 ROE'] = roe_melted.groupby('CIK')['ROE'].shift(1)
roe_melted['Y+1 ROE'] = roe_melted.groupby('CIK')['ROE'].shift(-1)
```

```python
import re

# Assuming 'TICKER FROM BLOOMBERG' is the name of your column
financial_data['Ticker'] = financial_data['Ticker'].apply(lambda x: re.sub(r'\s*US.*$', ' US', x))
```

```python
financial_data.head(5)
```

```python
nan_values = financial_data.isna().sum()

# Display the count of NaN values for each column
print(nan_values)
```

```python
merged_df = pd.merge(roe_melted, financial_data, how='inner', left_on=['CIK', 'Year'], right_on=['CIK', 'Year'])
merged_df = pd.merge(merged_df, asset_melted, how='inner', on=['CIK','Year'])
```

```python
merged_df.drop(columns=['Ticker_y','Ticker'],inplace=True)
```

```python
merged_df.rename(columns={ 'Ticker_x':'Ticker'},inplace=True)
```

```python
desired_columns = ['Ticker', 'CIK', 'Year','Y-1 ROE', 'ROE','Y+1 ROE', 'Y-1 Asset','Asset', 'Y+1 Asset','Year-1 Revenue',
       'Revenue', 'Year+1 Revenue', 'Year-1 Net Income', 'Net Income',
       'Year+1 Net Income','Accession Number', 'Conformed Submission Type',
       'Conformed Period Of Report', 'Filed As Of Date', 'Date As Of Change',
       'Company Conformed Name', 'Fiscal Year End', 'Y-1 Profit Margin',
       'Y Profit Margin', 'Y+1 Profit Margin', 'Current Revenue YOY',
       'Next Revenue YOY', 'Current Profit Margin YOY',
       'Next Profit Margin YOY']
```

```python
merged_df = merged_df[desired_columns]
merged_df
```

```python
nan_values = merged_df.isna().sum()

# Display the count of NaN values for each column
print(nan_values)
```

```python
# Group by CIK
grouped = merged_df.groupby('CIK')

# Define a function to calculate Current and Future ROE YOY
def calculate_roe_yoy(group):
    # Calculate Current ROE YOY
    group['Current ROE YOY'] = group['ROE'].astype(float) - group['Y-1 ROE'].astype(float)
    
    # Calculate Future ROE YOY
    group['Future ROE YOY'] = group['Y+1 ROE'].astype(float) - group['ROE'].astype(float)
    
    return group

# Apply the function to each group
merged_df = grouped.apply(calculate_roe_yoy)

# Reset index if needed
merged_df.reset_index(drop=True, inplace=True)

```

```python
merged_df
```

```python
merged_df.columns
```

```python
nan_values = merged_df.isna().sum()

# Display the count of NaN values for each column
print(nan_values)
```

```python
# Fill NaN values with 0 in specific columns
merged_df[['Y-1 ROE', 'Y+1 ROE','Y-1 Asset', 'Y+1 Asset']] = merged_df[['Y-1 ROE', 'Y+1 ROE','Y-1 Asset', 'Y+1 Asset']].fillna(0)
```
