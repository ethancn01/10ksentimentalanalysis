```python
#Import required libraries
import pandas as pd
import numpy as np
```

```python
#Read the Excel file into pandas DataFrame
revenue_df = pd.read_excel('Bloomberg_revenue_income.xlsx', sheet_name='revenue')
income_df = pd.read_excel('Bloomberg_revenue_income.xlsx', sheet_name='income')
```

```python
#Extracting year columns
year_columns_revenues = revenue_df.columns[3:]
year_columns_income = income_df.columns[3:]
```

```python
# Melt the DataFrame to bring all years into a single column,
# while keeping tickers, Ticker from Bloomberg, and CIK from Bloomberg as identifiers.

revenue_melted = pd.melt(revenue_df, id_vars=['tickers', 'Ticker from Bloomberg', 'CIK from Bloomberg'], value_vars=year_columns_revenues, var_name='Year', value_name='Revenue')
income_melted = pd.melt(income_df, id_vars=['tickers', 'Ticker from Bloomberg', 'CIK from Bloomberg'], value_vars=year_columns_income, var_name='Year', value_name='Income')

# Sort the melted DataFrame first by tickers and then by year

revenue_melted_sorted = revenue_melted.sort_values(by=['tickers', 'Year'])
income_melted_sorted = income_melted.sort_values(by=['tickers', 'Year'])
```

```python
# Merge the two DataFrames on both 'tickers' and 'year'
merged_df = pd.merge(revenue_melted_sorted, income_melted_sorted, on=['tickers', 'Year'])

# Display the merged DataFrame
merged_df
```

```python
#Clean the merged dataframe

# identical_columns = merged_df['Ticker from Bloomberg_x'].equals(merged_df['Ticker from Bloomberg_y'])
#identical_columns
```

```python
# Check if the two columns are identical
identical_columns = merged_df['Ticker from Bloomberg_x'].equals(merged_df['Ticker from Bloomberg_y'])

# If the columns are identical, drop one of them
if identical_columns:
    merged_df.drop(columns=['Ticker from Bloomberg_y'], inplace=True)
    merged_df.rename(columns={'Ticker from Bloomberg_x': 'Ticker from Bloomberg'}, inplace=True)
# Print the result

```

```python
#combined_data.replace(' ', np.nan, inplace=True)

# # Calculate Profit Margin
# combined_data['Y-1 PROFIT MARGIN'] = combined_data['YEAR-1 NET INCOME'].astype(float) / combined_data['YEAR-1 REVENUE'].astype(float)
# combined_data['Y PROFIT MARGIN'] = combined_data['NET INCOME'].astype(float) / combined_data['REVENUE'].astype(float)
# combined_data['Y+1 PROFIT MARGIN'] = combined_data['YEAR+1 NET INCOME'].astype(float) / combined_data['YEAR+1 REVENUE'].astype(float)



# Calculate Profit Margin for each CIK
combined_data['Y-1 Profit Margin'] = combined_data.groupby('CIK')['Year-1 Net Income'].transform(lambda x: x.astype(float) / combined_data['Year-1 Revenue'].astype(float))
combined_data['Y Profit Margin'] = combined_data.groupby('CIK')['Net Income'].transform(lambda x: x.astype(float) / combined_data['Revenue'].astype(float))
combined_data['Y+1 Profit Margin'] = combined_data.groupby('CIK')['Year+1 Net Income'].transform(lambda x: x.astype(float) / combined_data['Year+1 Revenue'].astype(float))
```

```python
combined_data
```

```python
# # Calculate YOY
# combined_data['CURRENT REVENUE YOY'] = (combined_data['REVENUE'].astype(float) - combined_data['YEAR-1 REVENUE'].astype(float)) / combined_data['YEAR-1 REVENUE'].astype(float)
# combined_data['NEXT REVENUE YOY'] = (combined_data['YEAR+1 REVENUE'].astype(float) - combined_data['REVENUE'].astype(float)) / combined_data['REVENUE'].astype(float)
# combined_data['CURRENT PROFIT MARGIN YOY'] = combined_data['Y PROFIT MARGIN'] - combined_data['Y-1 PROFIT MARGIN']
# combined_data['NEXT PROFIT MARGIN YOY'] = combined_data['Y+1 PROFIT MARGIN'] - combined_data['Y PROFIT MARGIN']

# Calculate CURRENT REVENUE YOY for each CIK
combined_data['Current Revenue YOY'] = combined_data.groupby('CIK').apply(lambda group: (group['Revenue'].astype(float) - group['Year-1 Revenue'].astype(float)) / group['Year-1 Revenue'].astype(float)).reset_index(drop=True)

# Calculate NEXT REVENUE YOY for each CIK
combined_data['Next Revenue YOY'] = combined_data.groupby('CIK').apply(lambda group: (group['Year+1 Revenue'].astype(float) - group['Revenue'].astype(float)) / group['Revenue'].astype(float)).reset_index(drop=True)

# Calculate CURRENT PROFIT MARGIN YOY for each CIK
combined_data['Current Profit Margin YOY'] = combined_data.groupby('CIK')['Y Profit Margin'].transform(lambda x: x - x.shift(1))

# Calculate NEXT PROFIT MARGIN YOY for each CIK
combined_data['Next Profit Margin YOY'] = combined_data.groupby('CIK')['Y+1 Profit Margin'].transform(lambda x: x - combined_data['Y Profit Margin'])
```
